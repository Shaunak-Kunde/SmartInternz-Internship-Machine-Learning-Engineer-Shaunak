```python
from sklearn.preprocessing import LabelEncoder

class MultiColumnLabelEncoder:
    def __init__(self, columns=None):
        self.columns = columns

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        output = X.copy()
        if self.columns is not None:
            for col in self.columns:
                output[col] = LabelEncoder().fit_transform(output[col])
        else:
            for colname in output.select_dtypes(include=['object']).columns:
                output[colname] = LabelEncoder().fit_transform(output[colname])
        return output

    def fit_transform(self, X, y=None):
        return self.fit(X, y).transform(X)
```

```python
Mcle = MultiColumnLabelEncoder()
data = Mcle.fit_transform(data)
```

```python
x=data.drop(['actual_productivity'],axis=1)
y=data['actual_productivity']
```

```python
X=x.to_numpy()
X
```

```python
#split and train
x_train,x_test,y_train,y_test=train_test_split(X,y,train_size=0.8,random_state=0)
```

```python
#linear regression model
model_lr=LinearRegression()
model_lr.fit(x_train, y_train)
```

```python
pred_test=model_lr.predict(x_test)
print("test MSE:",mean_squared_error(y_test,pred_test))
print("test MAE:",mean_absolute_error(y_test,pred_test))
print("R2 Score:{}".format(r2_score(y_test,pred_test)))
```

```python
#random forest regressor
model_rf=RandomForestRegressor(n_estimators=200,max_depth=5)
model_rf.fit(x_train, y_train)
```

```python
pred=model_rf.predict(x_test)
print("test MSE:",mean_squared_error(y_test,pred))
print("test MAE:",mean_absolute_error(y_test,pred))
print("R2 Score:{}".format(r2_score(y_test,pred)))
```

```python
#model xgb
model_xgb=xgb.XGBRegressor(n_estimators=200,max_depth=5,learning_rate=0.1)
model_xgb.fit(x_train, y_train)
```

```python
pred3=model_xgb.predict(x_test)
print("test MSE:",mean_squared_error(y_test,pred3))
print("test MAE:",mean_absolute_error(y_test,pred3))
print("R2 Score:{}".format(r2_score(y_test,pred3)))
```
